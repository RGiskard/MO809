```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_palette("Set2")
```

```python
df = pd.read_csv('SGD_Normal_LR.csv')
df.head(10)
```

```python
df['Remaining Weights'] = 1 - df['prune_proportion']
```

```python
g = sns.FacetGrid(data=df, col='dataset', hue='initialisation', height=4, sharex=False, sharey=False, col_wrap=2, aspect=1.5)

g.map(sns.lineplot, 'Remaining Weights', 'accuracy').set(xscale='log')
for ax in g.axes:
    ax.invert_xaxis()
```
